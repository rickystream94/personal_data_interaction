```python
import numpy as np
import random as rn
from datetime import date
import datetime
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
```

```python
values1 = [rn.randint(-3,3) for _ in range(30)]
values2 = [rn.randint(-3,3) for _ in range(30)]
```

```python
base = date.today()
date_list = [base - datetime.timedelta(days=x) for x in range(30)]
mpl_data = mdates.date2num(date_list)

fig, ax = plt.subplots(1,1)
ax.patch.set_facecolor('lightyellow')
fig.set_size_inches(24.5, 16.5)
ax.plot(mpl_data,values1,'go-',linewidth=2)
ax.plot(mpl_data,values2,'ro-',linewidth=2)
locator = mdates.AutoDateLocator()
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(mdates.AutoDateFormatter(locator))
ax.axes.set_ylim(-5,5)
ax.set_yticks([-3,-2,-1,0,1,2,3])
ax.set_xlabel("Time",fontsize=20)
ax.set_ylabel("EEG Level",fontsize=20)
plt.grid()
plt.show()
fig.savefig('plot.png')
```
